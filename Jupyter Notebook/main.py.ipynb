```python
import pandas as pd
```

```python
class Model:

	def __init__(self,dataset):
		self._dataset = dataset
		self._data = pd.read_csv(self._dataset)

	def checkData(self):
		print(self._data.head())
```

```python
dataset = "Simulated_data_ageing.csv"
model = Model(dataset)
df = model._data
```

```python
print(df.head())
```

```python
df['A_mean'] = df[['A_1','A_2','A_3','A_3']].mean(axis=1)
df['B_mean'] = df[['B_1','B_2','B_3','B_3']].mean(axis=1)
df['C_mean'] = df[['C_1','C_2','C_3','C_3']].mean(axis=1)
df['D_mean'] = df[['D_1','D_2','D_3','D_3']].mean(axis=1)
```

```python
df
```

```python
df['A_SD'] = df[['A_1','A_2','A_3','A_3']].std(axis=1)
df['B_SD'] = df[['B_1','B_2','B_3','B_3']].std(axis=1)
df['C_SD'] = df[['C_1','C_2','C_3','C_3']].std(axis=1)
df['D_SD'] = df[['D_1','D_2','D_3','D_3']].std(axis=1)
```

```python
df.head()
```

```python
df['A_variation'] = df['A_SD']/df['A_mean']
df['B_variation'] = df['B_SD']/df['B_mean']
df['C_variation'] = df['C_SD']/df['C_mean']
df['D_variation'] = df['D_SD']/df['D_mean']
```
